```python
text_data = """
Artificial intelligence is transforming technology.
Artificial intelligence improves healthcare.
Machine learning is a part of artificial intelligence.
Deep learning is a subset of machine learning.
Healthcare technology is advancing rapidly.
"""
```

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text.split()

words = preprocess_text(text_data)
```

```python
from collections import defaultdict

markov_chain = defaultdict(list)

for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]
    markov_chain[current_word].append(next_word)
```

```python
{
  "artificial": ["intelligence", "intelligence"],
  "intelligence": ["is", "improves"]
}
```

```python
import random

def generate_text(chain, start_word, length=20):
    current_word = start_word
    generated_words = [current_word]

    for _ in range(length):
        next_words = chain.get(current_word)
        if not next_words:
            break
        current_word = random.choice(next_words)
        generated_words.append(current_word)

    return " ".join(generated_words)
```

```python
start = "artificial"
output_text = generate_text(markov_chain, start_word=start, length=15)

print("Generated Text:")
print(output_text)
```
